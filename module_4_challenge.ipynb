## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['line_subtotal'] = df['unit_price']*df['qty']
# df.loc[df['unit_price','subcategory', 'line_subtotal' ]].head()
df['line_subtotal'].head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def calculate_shipping_price(weight):
    if weight > 50:
        # $7 per pound for orders over 50 pounds
        return weight * 7
    else:
        # $10 per pound for items 50 pounds or under
        return weight * 10
df['shipping_price'] = df['unit_weight'].apply(calculate_shipping_price)*df['qty']
df['shipping_price'].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
TAX_RATE = 0.0925
df['total_price'] = df['line_subtotal'] + df['shipping_price'] + (df['line_subtotal'] + df['shipping_price']) * TAX_RATE
df['total_price'].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_price'] = df['unit_price'] * df['qty'] + df['shipping_price']
df['line_price'].head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['profit'] = df['total_price'] - df['line_price']
df['profit'].sort_values(ascending=False).head()
#df['profit'].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
# This is to calculate the quantity purchased by each of the top 5 clients
top_clients_by_quantity = df.groupby('client_id')['qty'].sum().nlargest(5)
top_clients_by_quantity.head()

# top_clients_data.head(5)
```

```python
# This is to filter the original DataFrame to include only transactions from these top clients
top_clients_data = df[df['client_id'].isin(top_clients_by_quantity.index)]
top_clients_data.head(5)
```

```python
# This is calculate the total amount spent by quantity for each top client
amount_spent_by_client = top_clients_data.groupby('client_id')['qty'].sum().sort_values(ascending=False)
# This is to print out the top 5 results
print("Amount spent by each of the top 5 clients by quantity:")
amount_spent_by_client.head()
```

```python
# This is to summarize the DataFrame showing the totals for the
# top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit.
summary_df = top_clients_data.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()
# This is to display thdefault top 5 results of the summary
summary_df.head()


```

```python
# This is to format the
# Currency in millions of dollars.
# Function
def to_millions(x):
    return '${:,.2f}M'.format(x / 1000000)
# Application of the function on the total revenue
summary_df['total_revenue'] = pd.to_numeric(summary_df['total_revenue'])
summary_df['total_revenue'] = summary_df['total_revenue'].apply(to_millions).sort_values(ascending=False)
summary_df['total_revenue'].head()
```

```python
# Application of the function on the total profit
summary_df['total_profit'] = summary_df['total_profit'].apply(to_millions).sort_values(ascending=False)
summary_df['total_profit'].head()
```

```python
# This is to rename the columns
# to names suitable for presentation.
summary_df = summary_df.rename(columns={
    'client_id': 'Client ID',
    'total_units_purchased': 'Total Units Purchased',
    'total_shipping_price': 'Total Shipping Price',
    'total_revenue': 'Total Revenue (Millions)',
    'total_profit': 'Total Profit (Millions)'
})
# This is to display the summary datra frame after renaming columns 
# to names suitable for presentation
sorted_summary_df=summary_df.sort_values(by='Total Revenue (Millions)', ascending=False)
sorted_summary_df.head()
```

```python
# This is to sort the updated data by "Total Profit" from highest to lowest
summary_df.sort_values(by='Total Profit (Millions)', ascending=False)
summary_df.head()
```
