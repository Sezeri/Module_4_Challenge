## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['line_subtotal'] = df['unit_price']*df['qty']
# df.loc[df['unit_price','subcategory', 'line_subtotal' ]].head()
df['line_subtotal'].head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def calculate_shipping_price(weight):
    if weight > 50:
        # $7 per pound for orders over 50 pounds
        return weight * 7
    else:
        # $10 per pound for items 50 pounds or under
        return weight * 10
df['total_weight'] = df['unit_weight']*df['qty']
df['shipping_price'] = df['total_weight'].apply(calculate_shipping_price)
df['shipping_price'].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
TAX_RATE = 0.0925
df['total_price'] = df['line_subtotal'] + df['shipping_price'] + (df['line_subtotal'] + df['shipping_price']) * TAX_RATE
df['total_price'].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_price'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['profit'] = df['total_price'] - df['line_price']
#df['profit'].sort_values(ascending=False).head()
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
# This is to calculate the quantity purchased by each of the top 5 clients
top_clients_by_quantity = list(df['client_id'].value_counts().head(5).index)
top_clients_by_quantity


# top_clients_data.head(5)
```

```python
def client_total(client_id, column):
    client_df = df.loc[df['client_id']== client_id, column]
    return client_df.sum()

for client_id in top_clients_by_quantity:
    print(f"{client_id}:${client_total(client_id,'total_price')}")
    
```

```python
top_clients_data ={'client_id': top_clients_by_quantity}
columns = ['qty', 'shipping_price', 'total_price', 'line_price','profit']

for col in columns:
    top_clients_data[col] = [client_total(client_id,col) for client_id in top_clients_by_quantity]

summary_df = pd.DataFrame(top_clients_data)
summary_df
```

```python
# This is to format the
# Currency in millions of dollars.
# Function
def to_millions(x):
    return '${:,.2f}M'.format(x / 1000000)
# Application of the function on the total revenue or profit
summary_df['profit'] = pd.to_numeric(summary_df['profit'])
summary_df['profit'] = summary_df['profit'].apply(to_millions).sort_values(ascending=False)
summary_df['profit'].head()
```

```python
# This is to rename the columns
# to names suitable for presentation.
summary_df = summary_df.rename(columns={
    'client_id': 'Client ID',
    'total_units_purchased': 'Units Purchased',
    'total_shipping_price': 'Total Shipping Price',
    'total_price': 'Total Price (Millions)',
    'line_price': 'Line Price (Millions)',
    'profit': 'Profit (Millions)'
    
})
# This is to display the summary data frame after renaming columns 
# to names suitable for presentation
# sorted_summary_df=summary_df.sort_values(by='Total Price (Millions)', ascending=False)
summary_df.head()
```
